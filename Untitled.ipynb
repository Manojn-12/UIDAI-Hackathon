```python
# UIDAI Aadhaar Analytics
# Environment configuration and library imports

import os
import glob
import warnings
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Suppress non-critical warnings
warnings.filterwarnings("ignore")

# Pandas display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", 50)

# Visualization defaults
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("husl")

plt.rcParams["figure.figsize"] = (14, 7)
plt.rcParams["font.size"] = 11
plt.rcParams["axes.titlesize"] = 13
plt.rcParams["axes.labelsize"] = 11
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10
plt.rcParams["legend.fontsize"] = 10
```

```python
def load_and_merge_csv_files(file_pattern, dataset_label):
    """
    Load multiple CSV files matching a pattern and merge them into a single DataFrame.
    """
    files = sorted(glob.glob(file_pattern))

    if not files:
        raise FileNotFoundError(f"No files found for pattern: {file_pattern}")

    print(f"\n{dataset_label} Dataset")
    print(f"Files detected: {len(files)}")

    dataframes = []
    for filepath in files:
        df_part = pd.read_csv(filepath)
        dataframes.append(df_part)
        print(f"  Loaded {os.path.basename(filepath)} → {len(df_part):,} rows")

    merged_df = pd.concat(dataframes, ignore_index=True)
    print(f"Total records after merge: {len(merged_df):,}")

    return merged_df


def load_all_datasets(data_dir="Data"):
    """
    Load Aadhaar enrolment, demographic update, and biometric update datasets.
    """

    enrolment_files = os.path.join(
        data_dir, "api_data_aadhar_enrolment", "api_data_aadhar_enrolment_*.csv"
    )

    demographic_files = os.path.join(
        data_dir, "api_data_aadhar_demographic", "api_data_aadhar_demographic_*.csv"
    )

    biometric_files = os.path.join(
        data_dir, "api_data_aadhar_biometric", "api_data_aadhar_biometric_*.csv"
    )

    enrolment_df = load_and_merge_csv_files(enrolment_files, "Enrolment")
    demographic_df = load_and_merge_csv_files(demographic_files, "Demographic Updates")
    biometric_df = load_and_merge_csv_files(biometric_files, "Biometric Updates")

    return enrolment_df, demographic_df, biometric_df


# Load datasets
enrolment_df, demographic_df, biometric_df = load_all_datasets()
```

```python
def preprocess_enrolment_data(df):
    """
    Clean and prepare Aadhaar enrolment dataset.
    """

    # Date parsing
    df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y", errors="coerce")

    # Temporal features
    df["year"] = df["date"].dt.year
    df["month"] = df["date"].dt.month
    df["quarter"] = df["date"].dt.quarter
    df["month_year"] = df["date"].dt.to_period("M")
    df["day_of_week"] = df["date"].dt.dayofweek

    # Handle missing numeric values
    enrolment_cols = ["age_0_5", "age_5_17", "age_18_greater"]
    df[enrolment_cols] = df[enrolment_cols].fillna(0)

    # Total enrolments
    df["total_enrolments"] = (
        df["age_0_5"] + df["age_5_17"] + df["age_18_greater"]
    )

    # Remove duplicates
    initial_rows = len(df)
    df = df.drop_duplicates()

    print(
        f"Enrolment dataset: {len(df):,} records "
        f"({initial_rows - len(df):,} duplicates removed)"
    )

    return df


def preprocess_demographic_data(df):
    """
    Clean and prepare Aadhaar demographic update dataset.
    """

    # Date parsing
    df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y", errors="coerce")

    # Temporal features
    df["year"] = df["date"].dt.year
    df["month"] = df["date"].dt.month
    df["quarter"] = df["date"].dt.quarter
    df["month_year"] = df["date"].dt.to_period("M")
    df["day_of_week"] = df["date"].dt.dayofweek

    # Handle missing numeric values
    update_cols = ["demo_age_5_17", "demo_age_17_"]
    df[update_cols] = df[update_cols].fillna(0)

    # Standardize column naming
    df.rename(columns={"demo_age_17_": "demo_age_17_plus"}, inplace=True)

    # Total demographic updates
    df["total_demo_updates"] = (
        df["demo_age_5_17"] + df["demo_age_17_plus"]
    )

    # Remove duplicates
    initial_rows = len(df)
    df = df.drop_duplicates()

    print(
        f"Demographic dataset: {len(df):,} records "
        f"({initial_rows - len(df):,} duplicates removed)"
    )

    return df


def preprocess_biometric_data(df):
    """
    Clean and prepare Aadhaar biometric update dataset.
    """

    # Date parsing
    df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y", errors="coerce")

    # Temporal features
    df["year"] = df["date"].dt.year
    df["month"] = df["date"].dt.month
    df["quarter"] = df["date"].dt.quarter
    df["month_year"] = df["date"].dt.to_period("M")
    df["day_of_week"] = df["date"].dt.dayofweek

    # Handle missing numeric values
    update_cols = ["bio_age_5_17", "bio_age_17_"]
    df[update_cols] = df[update_cols].fillna(0)

    # Standardize column naming
    df.rename(columns={"bio_age_17_": "bio_age_17_plus"}, inplace=True)

    # Total biometric updates
    df["total_bio_updates"] = (
        df["bio_age_5_17"] + df["bio_age_17_plus"]
    )

    # Remove duplicates
    initial_rows = len(df)
    df = df.drop_duplicates()

    print(
        f"Biometric dataset: {len(df):,} records "
        f"({initial_rows - len(df):,} duplicates removed)"
    )

    return df


# Apply preprocessing
enrolment_df = preprocess_enrolment_data(enrolment_df)
demographic_df = preprocess_demographic_data(demographic_df)
biometric_df = preprocess_biometric_data(biometric_df)
```

```python
# State and location standardization mappings

STATE_MAPPING = {
    # West Bengal
    "West Bengal": "West Bengal",
    "WEST BENGAL": "West Bengal",
    "West bengal": "West Bengal",
    "west Bengal": "West Bengal",
    "Westbengal": "West Bengal",
    "WESTBENGAL": "West Bengal",
    "West  Bengal": "West Bengal",
    "West Bangal": "West Bengal",
    "West Bengli": "West Bengal",

    # Odisha
    "Odisha": "Odisha",
    "ODISHA": "Odisha",
    "odisha": "Odisha",
    "Orissa": "Odisha",

    # Andhra Pradesh
    "Andhra Pradesh": "Andhra Pradesh",
    "andhra pradesh": "Andhra Pradesh",

    # Jammu and Kashmir
    "Jammu and Kashmir": "Jammu and Kashmir",
    "Jammu And Kashmir": "Jammu and Kashmir",
    "Jammu & Kashmir": "Jammu and Kashmir",

    # Chhattisgarh
    "Chhattisgarh": "Chhattisgarh",
    "Chhatisgarh": "Chhattisgarh",

    # Puducherry
    "Puducherry": "Puducherry",
    "Pondicherry": "Puducherry",

    # Uttarakhand
    "Uttarakhand": "Uttarakhand",
    "Uttaranchal": "Uttarakhand",

    # Tamil Nadu
    "Tamil Nadu": "Tamil Nadu",
    "Tamilnadu": "Tamil Nadu",

    # Andaman and Nicobar Islands
    "Andaman and Nicobar Islands": "Andaman and Nicobar Islands",
    "Andaman & Nicobar Islands": "Andaman and Nicobar Islands",

    # Dadra and Nagar Haveli and Daman and Diu
    "Dadra and Nagar Haveli and Daman and Diu": "Dadra and Nagar Haveli and Daman and Diu",
    "Dadra and Nagar Haveli": "Dadra and Nagar Haveli and Daman and Diu",
    "Dadra & Nagar Haveli": "Dadra and Nagar Haveli and Daman and Diu",
    "Daman and Diu": "Dadra and Nagar Haveli and Daman and Diu",
    "Daman & Diu": "Dadra and Nagar Haveli and Daman and Diu",
    "The Dadra And Nagar Haveli And Daman And Diu": "Dadra and Nagar Haveli and Daman and Diu",
}

CITY_TO_STATE = {
    "BALANAGAR": "Telangana",
    "Darbhanga": "Bihar",
    "Jaipur": "Rajasthan",
    "Madanapalle": "Andhra Pradesh",
    "Nagpur": "Maharashtra",
    "Puttenahalli": "Karnataka",
    "Raja Annamalai Puram": "Tamil Nadu",
}

# Combined mapping
FULL_STATE_MAPPING = {**STATE_MAPPING, **CITY_TO_STATE}


def clean_state_names(df, dataset_name):
    """
    Standardize state names and remove invalid entries.
    """

    initial_state_count = df["state"].nunique()

    # Apply state and city-to-state mapping
    df["state"] = df["state"].map(lambda x: FULL_STATE_MAPPING.get(x, x))

    # Remove numeric-only invalid state values
    invalid_mask = df["state"].astype(str).str.match(r"^\d+$", na=False)
    df = df.loc[~invalid_mask]

    final_state_count = df["state"].nunique()

    print(
        f"{dataset_name} dataset: "
        f"{initial_state_count} → {final_state_count} standardized states"
    )

    return df


# Apply state cleaning
enrolment_df = clean_state_names(enrolment_df, "Enrolment")
demographic_df = clean_state_names(demographic_df, "Demographic")
biometric_df = clean_state_names(biometric_df, "Biometric")

print(f"Final unique states across datasets: {enrolment_df['state'].nunique()}")
```

```python
# Univariate analysis: age-wise distribution

# Enrolment totals
age_0_5_total = enrolment_df["age_0_5"].sum()
age_5_17_total = enrolment_df["age_5_17"].sum()
age_18_plus_total = enrolment_df["age_18_greater"].sum()
total_enrolments = age_0_5_total + age_5_17_total + age_18_plus_total

print("Enrolments by age group")
print(f"  0–5 years   : {age_0_5_total:12,} ({age_0_5_total / total_enrolments * 100:5.2f}%)")
print(f"  5–17 years  : {age_5_17_total:12,} ({age_5_17_total / total_enrolments * 100:5.2f}%)")
print(f"  18+ years   : {age_18_plus_total:12,} ({age_18_plus_total / total_enrolments * 100:5.2f}%)")
print(f"  Total       : {total_enrolments:12,}")

# Demographic update totals
demo_5_17_total = demographic_df["demo_age_5_17"].sum()
demo_17_plus_total = demographic_df["demo_age_17_plus"].sum()
total_demo_updates = demo_5_17_total + demo_17_plus_total

print("\nDemographic updates by age")
print(f"  5–17 years  : {demo_5_17_total:12,} ({demo_5_17_total / total_demo_updates * 100:5.2f}%)")
print(f"  17+ years   : {demo_17_plus_total:12,} ({demo_17_plus_total / total_demo_updates * 100:5.2f}%)")
print(f"  Total       : {total_demo_updates:12,}")

# Biometric update totals
bio_5_17_total = biometric_df["bio_age_5_17"].sum()
bio_17_plus_total = biometric_df["bio_age_17_plus"].sum()
total_bio_updates = bio_5_17_total + bio_17_plus_total

print("\nBiometric updates by age")
print(f"  5–17 years  : {bio_5_17_total:12,} ({bio_5_17_total / total_bio_updates * 100:5.2f}%)")
print(f"  17+ years   : {bio_17_plus_total:12,} ({bio_17_plus_total / total_bio_updates * 100:5.2f}%)")
print(f"  Total       : {total_bio_updates:12,}")

# Overall service volume
total_served = total_enrolments + total_demo_updates + total_bio_updates
update_to_enrolment_ratio = (total_demo_updates + total_bio_updates) / total_enrolments

print(f"\nTotal citizens served: {total_served:,}")
print(f"Update-to-enrolment ratio: {update_to_enrolment_ratio:.2f}x")

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Enrolments
enrol_age_labels = ["0–5 years", "5–17 years", "18+ years"]
enrol_values = [age_0_5_total, age_5_17_total, age_18_plus_total]

bars = axes[0].bar(
    enrol_age_labels,
    enrol_values,
    color=["#3498db", "#2ecc71", "#e74c3c"],
    edgecolor="black",
    linewidth=1.2,
)
axes[0].set_title("Enrolments by Age Group")
axes[0].set_ylabel("Total Enrolments")
axes[0].grid(axis="y", linestyle="--", alpha=0.3)

for bar, value in zip(bars, enrol_values):
    pct = value / total_enrolments * 100
    axes[0].text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{value / 1e6:.2f}M\n({pct:.1f}%)",
        ha="center",
        va="bottom",
        fontsize=9,
        fontweight="bold",
    )

# Demographic updates
update_age_labels = ["5–17 years", "17+ years"]
demo_values = [demo_5_17_total, demo_17_plus_total]

bars = axes[1].bar(
    update_age_labels,
    demo_values,
    color=["#2ecc71", "#e74c3c"],
    edgecolor="black",
    linewidth=1.2,
)
axes[1].set_title("Demographic Updates by Age")
axes[1].set_ylabel("Total Updates")
axes[1].grid(axis="y", linestyle="--", alpha=0.3)

for bar, value in zip(bars, demo_values):
    pct = value / total_demo_updates * 100
    axes[1].text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{value / 1e6:.1f}M\n({pct:.1f}%)",
        ha="center",
        va="bottom",
        fontsize=9,
        fontweight="bold",
    )

# Biometric updates
bio_values = [bio_5_17_total, bio_17_plus_total]

bars = axes[2].bar(
    update_age_labels,
    bio_values,
    color=["#2ecc71", "#e74c3c"],
    edgecolor="black",
    linewidth=1.2,
)
axes[2].set_title("Biometric Updates by Age")
axes[2].set_ylabel("Total Updates")
axes[2].grid(axis="y", linestyle="--", alpha=0.3)

for bar, value in zip(bars, bio_values):
    pct = value / total_bio_updates * 100
    axes[2].text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{value / 1e6:.1f}M\n({pct:.1f}%)",
        ha="center",
        va="bottom",
        fontsize=9,
        fontweight="bold",
    )

plt.tight_layout()
plt.savefig("01_univariate_age_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Bivariate analysis: temporal trends (monthly and quarterly)

# Monthly aggregation
enrolment_monthly = (
    enrolment_df.groupby("month", as_index=False)
    .agg(
        total_enrolments=("total_enrolments", "sum"),
        age_0_5=("age_0_5", "sum"),
        age_5_17=("age_5_17", "sum"),
        age_18_greater=("age_18_greater", "sum"),
    )
)

demographic_monthly = (
    demographic_df.groupby("month", as_index=False)
    .agg(total_demo_updates=("total_demo_updates", "sum"))
)

biometric_monthly = (
    biometric_df.groupby("month", as_index=False)
    .agg(total_bio_updates=("total_bio_updates", "sum"))
)

month_labels_map = {
    3: "Mar", 4: "Apr", 5: "May", 6: "Jun",
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct",
    11: "Nov", 12: "Dec"
}

print("Monthly enrolments")
for _, row in enrolment_monthly.iterrows():
    m = int(row["month"])
    print(
        f"  {month_labels_map[m]}: {row['total_enrolments']:10,.0f} "
        f"(0–5: {row['age_0_5']:9,.0f}, "
        f"5–17: {row['age_5_17']:9,.0f}, "
        f"18+: {row['age_18_greater']:7,.0f})"
    )

print("\nMonthly updates")
for m in sorted(demographic_monthly["month"].unique()):
    demo_val = demographic_monthly.loc[
        demographic_monthly["month"] == m, "total_demo_updates"
    ].values[0]
    bio_val = biometric_monthly.loc[
        biometric_monthly["month"] == m, "total_bio_updates"
    ].values[0]
    print(f"  {month_labels_map[m]}: Demo {demo_val:12,.0f}, Bio {bio_val:12,.0f}")

# Quarterly aggregation
enrolment_quarterly = enrolment_df.groupby("quarter")["total_enrolments"].sum()
demographic_quarterly = demographic_df.groupby("quarter")["total_demo_updates"].sum()
biometric_quarterly = biometric_df.groupby("quarter")["total_bio_updates"].sum()

print("\nQuarterly totals")
for q in sorted(enrolment_quarterly.index):
    print(
        f"  Q{int(q)}: "
        f"Enrol {enrolment_quarterly[q]:11,.0f}, "
        f"Demo {demographic_quarterly[q]:12,.0f}, "
        f"Bio {biometric_quarterly[q]:12,.0f}"
    )

# Visualization
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

months = sorted(enrolment_monthly["month"].unique())
month_labels = [month_labels_map[m] for m in months]
x = np.arange(len(months))
bar_width = 0.25

# Monthly enrolments by age group
axes[0].bar(
    x - bar_width,
    enrolment_monthly["age_0_5"].values,
    bar_width,
    label="Age 0–5",
    color="#3498db",
    edgecolor="black",
    linewidth=1,
)
axes[0].bar(
    x,
    enrolment_monthly["age_5_17"].values,
    bar_width,
    label="Age 5–17",
    color="#2ecc71",
    edgecolor="black",
    linewidth=1,
)
axes[0].bar(
    x + bar_width,
    enrolment_monthly["age_18_greater"].values,
    bar_width,
    label="Age 18+",
    color="#e74c3c",
    edgecolor="black",
    linewidth=1,
)

axes[0].set_xlabel("Month")
axes[0].set_ylabel("Total Enrolments")
axes[0].set_title("Monthly Enrolment Trends by Age Group")
axes[0].set_xticks(x)
axes[0].set_xticklabels(month_labels)
axes[0].legend()
axes[0].grid(axis="y", linestyle="--", alpha=0.3)

# Monthly updates comparison
axes[1].plot(
    x,
    demographic_monthly["total_demo_updates"].values,
    marker="o",
    linewidth=2.5,
    markersize=7,
    label="Demographic Updates",
    color="#9b59b6",
    markeredgecolor="black",
)
axes[1].plot(
    x,
    biometric_monthly["total_bio_updates"].values,
    marker="s",
    linewidth=2.5,
    markersize=7,
    label="Biometric Updates",
    color="#e67e22",
    markeredgecolor="black",
)

axes[1].set_xlabel("Month")
axes[1].set_ylabel("Total Updates")
axes[1].set_title("Monthly Update Trends")
axes[1].set_xticks(x)
axes[1].set_xticklabels(month_labels)
axes[1].legend()
axes[1].grid(axis="y", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.savefig("02_temporal_trends_monthly.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Bivariate analysis: geographic patterns (state-level)

# State-wise enrolments
state_enrolments = (
    enrolment_df.groupby("state", as_index=False)
    .agg(
        total_enrolments=("total_enrolments", "sum"),
        age_0_5=("age_0_5", "sum"),
        age_5_17=("age_5_17", "sum"),
        age_18_greater=("age_18_greater", "sum"),
    )
    .sort_values("total_enrolments", ascending=False)
)

print("Top 15 states by enrolments")
for idx, row in state_enrolments.head(15).iterrows():
    rank = state_enrolments.index.get_loc(idx) + 1
    print(
        f"  {rank:2d}. {row['state']:<28} {row['total_enrolments']:8,.0f} "
        f"(0–5: {row['age_0_5']:7,.0f}, "
        f"5–17: {row['age_5_17']:7,.0f}, "
        f"18+: {row['age_18_greater']:6,.0f})"
    )

top_5_share = (
    state_enrolments.head(5)["total_enrolments"].sum()
    / state_enrolments["total_enrolments"].sum()
    * 100
)
print(f"\nTop 5 states share: {top_5_share:.1f}%")

# State-wise updates
state_demographic = (
    demographic_df.groupby("state", as_index=False)
    .agg(total_demo_updates=("total_demo_updates", "sum"))
    .sort_values("total_demo_updates", ascending=False)
)

state_biometric = (
    biometric_df.groupby("state", as_index=False)
    .agg(total_bio_updates=("total_bio_updates", "sum"))
    .sort_values("total_bio_updates", ascending=False)
)

print("\nTop 15 states by demographic updates")
for idx, row in state_demographic.head(15).iterrows():
    rank = state_demographic.index.get_loc(idx) + 1
    print(f"  {rank:2d}. {row['state']:<28} {row['total_demo_updates']:12,.0f}")

print("\nTop 15 states by biometric updates")
for idx, row in state_biometric.head(15).iterrows():
    rank = state_biometric.index.get_loc(idx) + 1
    print(f"  {rank:2d}. {row['state']:<28} {row['total_bio_updates']:12,.0f}")

# Combined activity summary
state_summary = (
    state_enrolments[["state", "total_enrolments"]]
    .merge(state_demographic, on="state", how="left")
    .merge(state_biometric, on="state", how="left")
)

state_summary.fillna(0, inplace=True)

state_summary["total_activity"] = (
    state_summary["total_enrolments"]
    + state_summary["total_demo_updates"]
    + state_summary["total_bio_updates"]
)

state_summary = state_summary.sort_values("total_activity", ascending=False)

print("\nTop 10 states by total activity")
for _, row in state_summary.head(10).iterrows():
    demo_ratio = (
        row["total_demo_updates"] / row["total_enrolments"]
        if row["total_enrolments"] > 0
        else 0
    )
    bio_ratio = (
        row["total_bio_updates"] / row["total_enrolments"]
        if row["total_enrolments"] > 0
        else 0
    )
    print(
        f"  {row['state']:<22} "
        f"E:{row['total_enrolments']:8,.0f} "
        f"D:{row['total_demo_updates']:10,.0f} "
        f"B:{row['total_bio_updates']:10,.0f} "
        f"({demo_ratio:5.1f}x, {bio_ratio:5.1f}x)"
    )

# Visualization
fig, axes = plt.subplots(2, 1, figsize=(15, 11))

# Top 15 states by enrolments
top_15_enrol = state_enrolments.head(15)
states = top_15_enrol["state"].values
enrol_values = top_15_enrol["total_enrolments"].values

colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(states)))
bars = axes[0].barh(
    states,
    enrol_values,
    color=colors,
    edgecolor="black",
    linewidth=1,
)

axes[0].set_xlabel("Total Enrolments")
axes[0].set_title("Top 15 States by Enrolments")
axes[0].grid(axis="x", linestyle="--", alpha=0.3)
axes[0].invert_yaxis()

for bar in bars:
    width = bar.get_width()
    axes[0].text(
        width,
        bar.get_y() + bar.get_height() / 2,
        f"{width / 1e3:.0f}K",
        va="center",
        ha="left",
        fontsize=9,
        fontweight="bold",
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
    )

# Top 15 states: updates comparison
top_states = state_summary.head(15)["state"].values

demo_vals = [
    state_demographic.loc[state_demographic["state"] == s, "total_demo_updates"].values[0]
    if len(state_demographic[state_demographic["state"] == s]) > 0 else 0
    for s in top_states
]

bio_vals = [
    state_biometric.loc[state_biometric["state"] == s, "total_bio_updates"].values[0]
    if len(state_biometric[state_biometric["state"] == s]) > 0 else 0
    for s in top_states
]

x = np.arange(len(top_states))
bar_width = 0.35

axes[1].barh(
    x - bar_width / 2,
    demo_vals,
    bar_width,
    label="Demographic Updates",
    color="#9b59b6",
    edgecolor="black",
    linewidth=1,
)
axes[1].barh(
    x + bar_width / 2,
    bio_vals,
    bar_width,
    label="Biometric Updates",
    color="#e67e22",
    edgecolor="black",
    linewidth=1,
)

axes[1].set_xlabel("Total Updates")
axes[1].set_title("Top 15 States: Demographic vs Biometric Updates")
axes[1].set_yticks(x)
axes[1].set_yticklabels(top_states)
axes[1].legend()
axes[1].grid(axis="x", linestyle="--", alpha=0.3)
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig("03_geographic_state_analysis.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Trivariate analysis: state, time, and age interactions

month_names = {
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr",
    5: "May", 6: "Jun", 7: "Jul", 8: "Aug",
    9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
}

# State × Month enrolment variability
state_month_enrol = (
    enrolment_df
    .groupby(["state", "month"], as_index=False)
    .agg(total_enrolments=("total_enrolments", "sum"))
)

state_month_pivot = (
    state_month_enrol
    .pivot(index="state", columns="month", values="total_enrolments")
    .fillna(0)
)

state_variability = state_month_pivot.std(axis=1).sort_values(ascending=False)

print("Top 10 states by monthly enrolment variability")
for i, (state, std_val) in enumerate(state_variability.head(10).items(), 1):
    print(f"  {i:2d}. {state:<28} {std_val:10,.0f}")

# Age × State × Quarter enrolment patterns
state_quarter_age = (
    enrolment_df
    .groupby(["state", "quarter"], as_index=False)
    .agg(
        age_0_5=("age_0_5", "sum"),
        age_5_17=("age_5_17", "sum"),
        age_18_greater=("age_18_greater", "sum"),
    )
)

age_variability_records = []

for state in state_quarter_age["state"].unique():
    subset = state_quarter_age[state_quarter_age["state"] == state]
    totals = subset[["age_0_5", "age_5_17", "age_18_greater"]].sum(axis=1)
    infant_pct_std = ((subset["age_0_5"] / totals) * 100).std()
    age_variability_records.append(
        {
            "state": state,
            "infant_pct_variability": infant_pct_std
        }
    )

age_variability_df = (
    pd.DataFrame(age_variability_records)
    .sort_values("infant_pct_variability", ascending=False)
)

print("\nStates with highest quarterly age-share variability")
for _, row in age_variability_df.head(10).iterrows():
    print(f"  {row['state']:<28} {row['infant_pct_variability']:10.2f}%")

# Visualization: Monthly enrolment distribution heatmap
fig, ax = plt.subplots(figsize=(16, 6))

top_states = state_variability.head(15).index
heatmap_data = state_month_pivot.loc[top_states]
heatmap_pct = heatmap_data.div(heatmap_data.sum(axis=1), axis=0) * 100

months_sorted = sorted(heatmap_pct.columns)
month_labels = [month_names[int(m)] for m in months_sorted]

im = ax.imshow(heatmap_pct, cmap="YlOrRd", aspect="auto")

ax.set_yticks(range(len(top_states)))
ax.set_yticklabels(top_states, fontsize=9)
ax.set_xticks(range(len(month_labels)))
ax.set_xticklabels(month_labels)
ax.set_xlabel("Month")
ax.set_ylabel("State")
ax.set_title("Monthly Enrolment Distribution by State (% of Annual Total)")

cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Percentage of Annual Total", rotation=270, labelpad=20)

plt.tight_layout()
plt.savefig("04_trivariate_analysis.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Trend and Anomaly Detection
# Identifying statistical, temporal, and geographic irregularities

# Statistical outlier detection (Z-score)
def detect_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    z_scores = np.abs((df[column] - mean) / std)
    outliers = df[z_scores > threshold].copy()
    outliers['z_score'] = z_scores[z_scores > threshold]
    return outliers.sort_values('z_score', ascending=False)

enrol_outliers = detect_outliers(enrolment_df, 'total_enrolments', threshold=3)

print("Top anomalous enrolment records")
print("Date        State              District           Volume   Z-Score")
for _, row in enrol_outliers.head(10).iterrows():
    print(
        f"{row['date'].date()}  "
        f"{row['state']:<18} "
        f"{row['district']:<18} "
        f"{row['total_enrolments']:>7,.0f} "
        f"{row['z_score']:>7.2f}"
    )

demo_outliers = detect_outliers(demographic_df, 'total_demo_updates', threshold=3)

print("\nTop anomalous demographic update records")
print("Date        State              District           Volume   Z-Score")
for _, row in demo_outliers.head(10).iterrows():
    print(
        f"{row['date'].date()}  "
        f"{row['state']:<18} "
        f"{row['district']:<18} "
        f"{row['total_demo_updates']:>7,.0f} "
        f"{row['z_score']:>7.2f}"
    )

# Temporal trend analysis (linear regression)
daily_enrol = enrolment_df.groupby('date')['total_enrolments'].sum().reset_index()
daily_enrol['days_from_start'] = (
    daily_enrol['date'] - daily_enrol['date'].min()
).dt.days

X = daily_enrol['days_from_start'].values.reshape(-1, 1)
y = daily_enrol['total_enrolments'].values

model = LinearRegression()
model.fit(X, y)

trend_slope = model.coef_[0]
r_squared = model.score(X, y)

print("\nEnrolment trend metrics")
print(f"Slope      : {trend_slope:,.2f} per day")
print(f"R-squared  : {r_squared:.4f}")

daily_demo = demographic_df.groupby('date')['total_demo_updates'].sum().reset_index()
daily_demo['days_from_start'] = (
    daily_demo['date'] - daily_demo['date'].min()
).dt.days

X_demo = daily_demo['days_from_start'].values.reshape(-1, 1)
y_demo = daily_demo['total_demo_updates'].values

model_demo = LinearRegression()
model_demo.fit(X_demo, y_demo)

trend_slope_demo = model_demo.coef_[0]
r_squared_demo = model_demo.score(X_demo, y_demo)

print("\nDemographic update trend metrics")
print(f"Slope      : {trend_slope_demo:,.2f} per day")
print(f"R-squared  : {r_squared_demo:.4f}")

# Geographic anomalies (district vs state average)
district_summary = (
    enrolment_df
    .groupby(['state', 'district'])['total_enrolments']
    .sum()
    .reset_index()
)

state_avg = district_summary.groupby('state')['total_enrolments'].mean().to_dict()
district_summary['state_avg'] = district_summary['state'].map(state_avg)
district_summary['deviation_ratio'] = (
    district_summary['total_enrolments'] / district_summary['state_avg']
)

geographic_outliers = district_summary[
    district_summary['deviation_ratio'] > 3
].sort_values('deviation_ratio', ascending=False)

print("\nDistricts with >3x state average enrolments")
print("State              District           Enrolments  State Avg  Ratio")
for _, row in geographic_outliers.head(15).iterrows():
    print(
        f"{row['state']:<18} "
        f"{row['district']:<18} "
        f"{row['total_enrolments']:>10,.0f} "
        f"{row['state_avg']:>10,.0f} "
        f"{row['deviation_ratio']:>6.1f}x"
    )

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Time series with trend and anomalies
axes[0, 0].plot(
    daily_enrol['date'],
    daily_enrol['total_enrolments'],
    linewidth=1.5
)
axes[0, 0].plot(
    daily_enrol['date'],
    model.predict(X),
    linestyle='--',
    linewidth=2
)

# Aggregate anomaly points by date to align x and y
anomaly_daily = (
    enrol_outliers
    .groupby('date')['total_enrolments']
    .sum()
    .reset_index()
)

axes[0, 0].scatter(
    anomaly_daily['date'],
    anomaly_daily['total_enrolments'],
    s=80,
    zorder=5
)

axes[0, 0].set_title('Daily Enrolments with Trend and Anomalies')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Total Enrolments')
axes[0, 0].grid(alpha=0.3, linestyle='--')
axes[0, 0].tick_params(axis='x', rotation=45)

# Distribution with outlier threshold
axes[0, 1].hist(
    enrolment_df['total_enrolments'],
    bins=50,
    edgecolor='black',
    alpha=0.7
)

mean_val = enrolment_df['total_enrolments'].mean()
std_val = enrolment_df['total_enrolments'].std()

axes[0, 1].axvline(mean_val, linestyle='--', linewidth=2)
axes[0, 1].axvline(mean_val + 3 * std_val, linestyle='--', linewidth=2)
axes[0, 1].set_title('Enrolment Distribution with Outlier Threshold')
axes[0, 1].set_xlabel('Total Enrolments per Record')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(axis='y', alpha=0.3, linestyle='--')

# Geographic outliers
top_geo = geographic_outliers.head(15)
axes[1, 0].barh(
    range(len(top_geo)),
    top_geo['deviation_ratio'],
    edgecolor='black'
)
axes[1, 0].set_yticks(range(len(top_geo)))
axes[1, 0].set_yticklabels(
    [f"{r['district']}, {r['state']}" for _, r in top_geo.iterrows()],
    fontsize=9
)
axes[1, 0].axvline(3, linestyle='--', linewidth=2)
axes[1, 0].set_xlabel('Ratio to State Average')
axes[1, 0].set_title('Top Geographic Outlier Districts')
axes[1, 0].invert_yaxis()
axes[1, 0].grid(axis='x', alpha=0.3, linestyle='--')

# Monthly anomaly heatmap
monthly_state = (
    enrolment_df
    .groupby(['state', 'month'])['total_enrolments']
    .sum()
    .reset_index()
)

pivot = monthly_state.pivot(
    index='state',
    columns='month',
    values='total_enrolments'
).fillna(0)

zscore_pivot = pivot.apply(
    lambda x: (x - x.mean()) / x.std() if x.std() > 0 else 0,
    axis=1
)

im = axes[1, 1].imshow(
    zscore_pivot.iloc[:20],
    aspect='auto',
    cmap='RdYlGn_r',
    vmin=-2,
    vmax=2
)

axes[1, 1].set_yticks(range(20))
axes[1, 1].set_yticklabels(zscore_pivot.index[:20], fontsize=8)
axes[1, 1].set_xticks(range(len(month_labels)))
axes[1, 1].set_xticklabels(month_labels)
axes[1, 1].set_title('Monthly Activity Z-Scores (Top 20 States)')
axes[1, 1].set_xlabel('Month')

plt.colorbar(im, ax=axes[1, 1], label='Z-Score')

plt.tight_layout()
plt.savefig('06_anomaly_detection.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ------------------------------------------------------------
# 7.1 Monthly Enrolment Forecast (Linear Trend)
# ------------------------------------------------------------

monthly_ts = enrolment_monthly.copy().reset_index(drop=True)
monthly_ts["month_index"] = range(len(monthly_ts))

X_forecast = monthly_ts["month_index"].values.reshape(-1, 1)
y_forecast = monthly_ts["total_enrolments"].values

forecast_model = LinearRegression()
forecast_model.fit(X_forecast, y_forecast)

future_months = np.array([[9], [10], [11]])
forecast_values = forecast_model.predict(future_months)
forecast_values = np.maximum(0, forecast_values)

forecast_df = pd.DataFrame({
    "month": ["Jan 2026", "Feb 2026", "Mar 2026"],
    "forecast_enrolments": forecast_values.astype(int)
})

r_squared_forecast = forecast_model.score(X_forecast, y_forecast)
mae_forecast = np.mean(np.abs(y_forecast - forecast_model.predict(X_forecast)))

print("\nForecasted Enrolments (Linear Projection)")
print(forecast_df)
print(f"\nModel R-squared : {r_squared_forecast:.4f}")
print(f"Model MAE       : {mae_forecast:,.0f}")

# ------------------------------------------------------------
# 7.2 Seasonal Pattern
# ------------------------------------------------------------

monthly_avg = enrolment_monthly.groupby("month")["total_enrolments"].mean()
monthly_pattern = (monthly_avg / monthly_avg.mean()) * 100

monthly_pattern_df = (
    monthly_pattern
    .reset_index()
    .rename(columns={"total_enrolments": "multiplier_pct"})
)

print("\nSeasonal Multipliers (% of Average Month)")
print(monthly_pattern_df.round(1))

# ------------------------------------------------------------
# 7.3 State-Level Growth Projection
# ------------------------------------------------------------

growth_records = []

for state in enrolment_df["state"].unique():
    state_series = (
        enrolment_df[enrolment_df["state"] == state]
        .groupby("month")["total_enrolments"]
        .sum()
        .sort_index()
    )

    if len(state_series) >= 3:
        growth_rate = state_series.pct_change().dropna().mean()
        current_val = state_series.iloc[-1]

        growth_records.append({
            "state": state,
            "growth_pct": round(growth_rate * 100, 1),
            "current_volume": int(current_val),
            "projected_next_month": int(current_val * (1 + growth_rate))
        })

growth_df = pd.DataFrame(growth_records).sort_values(
    "growth_pct", ascending=False
)

print("\nTop 10 States by Average Monthly Growth")
print(growth_df.head(10).reset_index(drop=True))

print("\nBottom 10 States by Average Monthly Growth")
print(growth_df.tail(10).reset_index(drop=True))

# ------------------------------------------------------------
# 7.4 Annual Volume & Capacity Projection
# ------------------------------------------------------------

months_observed = enrolment_monthly["month"].nunique()

avg_monthly_enrol = total_enrolments / months_observed
avg_monthly_demo = total_demo / months_observed
avg_monthly_bio = total_bio / months_observed

projected_2026_enrol = int(avg_monthly_enrol * 12)
projected_2026_demo = int(avg_monthly_demo * 12)
projected_2026_bio = int(avg_monthly_bio * 12)

projection_df = pd.DataFrame({
    "service": [
        "Enrolments",
        "Demographic Updates",
        "Biometric Updates"
    ],
    "projected_2026_volume": [
        projected_2026_enrol,
        projected_2026_demo,
        projected_2026_bio
    ]
})

print("\n2026 Annual Volume Projection")
print(projection_df)

avg_service_time_minutes = 15
operating_hours_per_day = 8
working_days_per_year = 300

services_per_center_per_day = (operating_hours_per_day * 60) / avg_service_time_minutes
total_services_2026 = projection_df["projected_2026_volume"].sum()

required_center_days = total_services_2026 / services_per_center_per_day
required_centers = required_center_days / working_days_per_year

print("\nCapacity Estimates")
print(f"Services per center/day : {int(services_per_center_per_day)}")
print(f"Center-days required    : {int(required_center_days):,}")
print(f"Estimated centers       : {int(required_centers):,}")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# ------------------------------------------------------------
# 1. Forecasted Enrolments (Next 3 Months)
# ------------------------------------------------------------

axes[0, 0].plot(
    forecast_df["month"],
    forecast_df["forecast_enrolments"],
    marker="o",
    linewidth=2
)
axes[0, 0].set_title("Enrolment Forecast (Next 3 Months)")
axes[0, 0].set_ylabel("Enrolments")
axes[0, 0].grid(axis="y", alpha=0.3)

for i, v in enumerate(forecast_df["forecast_enrolments"]):
    axes[0, 0].text(i, v, f"{v:,}", ha="center", va="bottom", fontsize=9)

# ------------------------------------------------------------
# 2. Seasonal Multipliers
# ------------------------------------------------------------

axes[0, 1].bar(
    monthly_pattern_df["month"],
    monthly_pattern_df["multiplier_pct"],
    edgecolor="black"
)
axes[0, 1].axhline(100, linestyle="--", linewidth=1)
axes[0, 1].set_title("Seasonal Multipliers (% of Average Month)")
axes[0, 1].set_ylabel("Multiplier (%)")
axes[0, 1].grid(axis="y", alpha=0.3)

# ------------------------------------------------------------
# 3. Top State Growth Rates
# ------------------------------------------------------------

top_growth = growth_df.head(10)

axes[1, 0].barh(
    top_growth["state"],
    top_growth["growth_pct"],
    edgecolor="black"
)
axes[1, 0].set_title("Top 10 States by Average Monthly Growth")
axes[1, 0].set_xlabel("Growth (%)")
axes[1, 0].invert_yaxis()
axes[1, 0].grid(axis="x", alpha=0.3)

# ------------------------------------------------------------
# 4. 2026 Projected Service Volumes
# ------------------------------------------------------------

axes[1, 1].bar(
    projection_df["service"],
    projection_df["projected_2026_volume"],
    edgecolor="black"
)
axes[1, 1].set_title("2026 Projected Annual Volumes")
axes[1, 1].set_ylabel("Volume")
axes[1, 1].grid(axis="y", alpha=0.3)
axes[1, 1].ticklabel_format(style="plain", axis="y")

for i, v in enumerate(projection_df["projected_2026_volume"]):
    axes[1, 1].text(i, v, f"{v/1e6:.1f}M", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.savefig("07_predictive_analysis.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Preconditions (hard fail if missing)
# ------------------------------------------------------------
required_vars = [
    "state_summary", "maturity_df", "monthly_pattern", "growth_df",
    "enrolment_df", "state_variability", "enrol_outliers",
    "month_names", "total_enrolments", "total_demo", "total_bio"
]

missing = [v for v in required_vars if v not in globals()]
if missing:
    raise NameError(f"Missing required variables: {missing}")

# ------------------------------------------------------------
# FRAMEWORK 1: STATE PRIORITIZATION MATRIX
# ------------------------------------------------------------
state_priority = (
    state_summary
    .merge(maturity_df[["state", "maturity_index"]], on="state", how="left")
)

state_priority["total_volume"] = (
    state_priority["total_enrolments"]
    + state_priority["total_demo_updates"]
    + state_priority["total_bio_updates"]
)

high_volume_threshold = state_priority["total_volume"].quantile(0.75)
high_maturity_threshold = 30

def classify_priority(row):
    if row["total_volume"] > high_volume_threshold and row["maturity_index"] > high_maturity_threshold:
        return "Priority 1"
    if row["total_volume"] > high_volume_threshold:
        return "Priority 2"
    if row["maturity_index"] > high_maturity_threshold:
        return "Priority 3"
    return "Priority 4"

state_priority["priority"] = state_priority.apply(classify_priority, axis=1)

# ------------------------------------------------------------
# FRAMEWORK 2: RESOURCE ALLOCATION
# ------------------------------------------------------------
national_centers = 15000
total_services = state_priority["total_volume"].sum()

state_priority["service_share_pct"] = state_priority["total_volume"] / total_services * 100
state_priority["recommended_centers"] = state_priority["service_share_pct"] / 100 * national_centers

state_priority["enrol_pct"] = state_priority["total_enrolments"] / state_priority["total_volume"] * 100
state_priority["demo_pct"] = state_priority["total_demo_updates"] / state_priority["total_volume"] * 100
state_priority["bio_pct"] = state_priority["total_bio_updates"] / state_priority["total_volume"] * 100

# ------------------------------------------------------------
# FRAMEWORK 3: TEMPORAL CAPACITY PLANNING
# ------------------------------------------------------------
baseline_staff = 100

monthly_staff = {
    m: baseline_staff * (v / 100)
    for m, v in monthly_pattern.items()
}

# ------------------------------------------------------------
# FRAMEWORK 4: PERFORMANCE BENCHMARKS
# ------------------------------------------------------------
national_maturity_avg = maturity_df["maturity_index"].mean()
national_bio_demo_ratio = total_bio / total_demo
national_enrol_per_district = total_enrolments / enrolment_df["district"].nunique()

# ------------------------------------------------------------
# FRAMEWORK 5: RISK & OPPORTUNITY FLAGS
# ------------------------------------------------------------
high_variability_states = state_variability.head(5).index.tolist()
declining_states = growth_df[growth_df["growth_pct"] < 0]["state"].tolist()
anomaly_states = enrol_outliers["state"].value_counts().head(5).index.tolist()

high_growth_states = growth_df.nlargest(5, "growth_pct")["state"].tolist()
underserved_states = state_priority.nsmallest(5, "total_volume")["state"].tolist()

# ------------------------------------------------------------
# VISUALIZATION DASHBOARD
# ------------------------------------------------------------
viz = state_priority.replace([np.inf, -np.inf], np.nan)
viz = viz.dropna(subset=["maturity_index", "total_volume"])

fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(3, 2, hspace=0.35, wspace=0.3)

# ---- Plot 1: Prioritization Matrix
ax1 = fig.add_subplot(gs[0, :])

priority_colors = {
    "Priority 1": "#e74c3c",
    "Priority 2": "#f39c12",
    "Priority 3": "#3498db",
    "Priority 4": "#95a5a6"
}

for p, c in priority_colors.items():
    d = viz[viz["priority"] == p]
    ax1.scatter(
        d["maturity_index"],
        d["total_volume"] / 1000,
        s=140,
        color=c,
        edgecolor="black",
        linewidth=1,
        alpha=0.85,
        label=p
    )

x = viz["maturity_index"].values
y = viz["total_volume"].values / 1000

if len(x) > 0:
    ax1.set_xlim(x.min() * 0.95, x.max() * 1.05)
if len(y) > 0:
    ax1.set_ylim(y.min() * 0.95, y.max() * 1.05)

ax1.axvline(viz["maturity_index"].median(), linestyle="--", linewidth=2)
ax1.axhline(viz["total_volume"].median() / 1000, linestyle="--", linewidth=2)

ax1.set_xlabel("Maturity Index")
ax1.set_ylabel("Total Service Volume (Thousands)")
ax1.set_title("State Prioritization Matrix")
ax1.legend(fontsize=9)
ax1.grid(alpha=0.3)

# ---- Plot 2: Resource Allocation
ax2 = fig.add_subplot(gs[1, 0])
top_alloc = viz.nlargest(15, "total_volume")

ax2.barh(
    top_alloc["state"],
    top_alloc["recommended_centers"],
    edgecolor="black"
)
ax2.set_xlabel("Recommended Centers")
ax2.set_title("Resource Allocation (Top 15 States)")
ax2.invert_yaxis()
ax2.grid(axis="x", alpha=0.3)

# ---- Plot 3: Monthly Staffing
ax3 = fig.add_subplot(gs[1, 1])

months_sorted = sorted(monthly_staff.keys())
staff_vals = [monthly_staff[m] for m in months_sorted]
month_labels = [month_names[m] for m in months_sorted]

ax3.bar(month_labels, staff_vals, edgecolor="black")
ax3.axhline(100, linestyle="--", linewidth=2)
ax3.set_ylabel("Staff per 100 Baseline")
ax3.set_title("Monthly Staffing Requirement")
ax3.grid(axis="y", alpha=0.3)

# ---- Plot 4: Risk & Opportunity Summary
ax4 = fig.add_subplot(gs[2, :])

risk_labels = ["High Variability", "Declining Trend", "Anomalies"]
risk_counts = [
    len(high_variability_states),
    len(declining_states),
    len(anomaly_states)
]

opp_labels = ["High Growth", "Underserved", "Weekend Expansion"]
opp_counts = [
    len(high_growth_states),
    len(underserved_states),
    len(state_priority)
]

x_risk = np.arange(len(risk_labels))
x_opp = np.arange(len(opp_labels)) + len(risk_labels) + 1

ax4.bar(x_risk, risk_counts, edgecolor="black", label="Risks")
ax4.bar(x_opp, opp_counts, edgecolor="black", label="Opportunities")

ax4.set_xticks(list(x_risk) + list(x_opp))
ax4.set_xticklabels(risk_labels + opp_labels, fontsize=9)
ax4.set_ylabel("States Affected")
ax4.set_title("Risk & Opportunity Dashboard")
ax4.legend()
ax4.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.savefig("08_decision_support_framework.png", dpi=300, bbox_inches="tight")
plt.show()
```
